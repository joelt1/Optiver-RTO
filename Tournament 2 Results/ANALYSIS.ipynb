```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### First Round

```python
df = pd.read_csv("match13_events.csv")
df.tail()
```

```python
team_data = df[df["Competitor"].str.contains("CashMoney")].dropna()
nptrivial_data = df[df["Competitor"].str.contains("NPTrivial")].dropna()
stonks_data = df[df["Competitor"].str.contains("StonksBot")].dropna()
# Profit and loss column for team
team_pl = team_data["ProfitLoss"]
nptrivial_pl = nptrivial_data["ProfitLoss"]
stonks_pl = stonks_data["ProfitLoss"]
len(team_data)
```

```python
plt.figure(figsize=(14, 9))
plt.plot(team_data["Time"], team_pl, label="CashMoney")
plt.plot(nptrivial_data["Time"], nptrivial_pl, label="NPTrivial")
plt.plot(stonks_data["Time"], stonks_pl, label="Stonks")
plt.xlabel("Time")
plt.ylabel("Profit or loss")
plt.title("Total profit and loss for each team in Round 1")
plt.legend(prop={'size': 12})
plt.tight_layout()
plt.show()
```
