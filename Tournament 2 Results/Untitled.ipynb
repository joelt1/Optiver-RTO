```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('match13_events.csv')
data.tail()
```

```python
traderone = data[data['Competitor'] == 'CashMoney']
```

```python
time = traderone["Time"]
PL = traderone["ProfitLoss"]
PL = 100*PL/max(abs(PL))
position = traderone["EtfPosition"]


fig, ax = plt.subplots(1,1, figsize = (14, 6))
ax.plot(time, position)
ax.plot(time, PL)

plt.legend(["Scaled position", "Scaled PL"])
```

```python
price = traderone["Price"]
etf_price = traderone["EtfPrice"]
time = traderone["Time"]
position = traderone["EtfPosition"]

filt = price > 3900
price = price[filt]
ptime = time[filt]
etf_price = etf_price[filt]
diff = price - etf_price

fig, ax = plt.subplots(1,1, figsize = (14, 6))
ax.plot(ptime, price)
ax.plot(ptime, etf_price)

plt.legend(["Price", "EtfPrice"])

fig2, ax2 = plt.subplots(1,1, figsize = (14, 6))
ax2.plot(time, position)

plt.legend(["Price", "EtfPrice"])

fig3, ax3 = plt.subplots(1,1, figsize = (14, 6))
ax3.plot(ptime, diff)

plt.legend(["Price", "EtfPrice"])
```
