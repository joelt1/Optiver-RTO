```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("match15_events.csv")
df.tail()
```

```python
team_data = df[df["Competitor"].str.contains("CashMoney")].dropna()
coolbeans_data = df[df["Competitor"].str.contains("CoolBeans")].dropna()
seabornpd_data = df[df["Competitor"].str.contains("SeabornPandas")].dropna()
# Profit and loss column for team
team_profit_loss = team_data["ProfitLoss"]
coolbeans_profit_loss = coolbeans_data["ProfitLoss"]
seabornpd_profit_loss = seabornpd_data["ProfitLoss"]
len(team_data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(team_data["Time"], team_profit_loss)
plt.xlabel("Time")
plt.ylabel("Profit or loss")
plt.title("Total profit and loss for team CashMoney")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(coolbeans_data["Time"], coolbeans_profit_loss)
plt.xlabel("Time")
plt.ylabel("Profit or loss")
plt.title("Total profit and loss for team CoolBeans")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(seabornpd_data["Time"], seabornpd_profit_loss)
plt.xlabel("Time")
plt.ylabel("Profit or loss")
plt.title("Total profit and loss for team SeabornPandas")
plt.tight_layout()
plt.show()
```
